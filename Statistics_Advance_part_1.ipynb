QUE 16  How do you calculate a Z-score, and what does it represent)

A Z-score, or standard score, is a way of quantifying how far a data point is from the mean of a dataset in terms of standard deviations. It's a helpful statistical tool for comparing data points from different datasets or distributions.
To calculate a Z-score, use this formula:
$$ Z = \frac{(X - \mu)}{\sigma} $$
Where:


What does it represent?
 1. Positive Z-score: Indicates the data point is above the mean.
 2.Negative Z-score: Indicates the data point is below the mean.
 3. Magnitude: The larger the absolute value of the Z-score, the farther the data point is from the mean. For example, a Z-score of +2 means the value is 2 standard deviations above the mean.

QUE 22 What is the concept of expected value in a probability distribution)

ANS The expected value (often denoted as E[X]) is a fundamental concept in probability and statistics that represents the average outcome you would expect from a random variable if you could repeat the experiment infinitely many times. It provides a measure of the "center" or "long-term average" of the probability distribution.

FORMULA
The expected value is calculated differently depending on whether the random variable is discrete or continuous:
Discrete Random Variable
If ( X ) is a discrete random variable with possible values ( x_1, x_2, ..., x_n ), and their corresponding probabilities ( P(x_1), P(x_2), ..., P(x_n) ), the expected value is:
$$ E[X] = \sum_{i=1}^{n} x_i \cdot P(x_i) $$

QUE 2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats # import the stats module from scipy

def plot_discrete_uniform_pmf():
    x = np.arange(1, 7)
    pmf = stats.randint.pmf(x, 1, 7)
    plt.bar(x, pmf, color='blue', alpha=0.7)
    plt.xlabel('Outcomes')
    plt.ylabel('Probability')
    plt.title('Discrete Uniform Distribution PMF')
    plt.show()

plot_discrete_uniform_pmf()
```

QUE 9 Write a Python function to calculate Z-scores from a dataset and plot them

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='cyan')
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Distribution")
    plt.show()
```

QUE 10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
def clt_simulation(sample_size=30, num_samples=1000):
    population = np.random.exponential(scale=2, size=10000)
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='brown')
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.title("Central Limit Theorem Simulation")
    plt.show()
```

QUE 14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution

```python
def hypothesis_test(sample_mean, pop_mean, pop_std, n):
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    print(f"Z-score: {z}, P-value: {p_value}")

hypothesis_test(52, 50, 10, 30)
```

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    print(f"Confidence Interval: {ci}")

confidence_interval(sample_data)
```

QUE 18 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the normal distribution
mean = 0
std_dev = 1

# Generate data
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Plot
plt.plot(x, pdf, label="PDF", color="blue")
plt.title("Probability Density Function (Normal Distribution)")
plt.xlabel("X")
plt.ylabel("Density")
plt.legend()
plt.show()
```

QUE 21 Write a Python program to compare the standard deviations of two datasets and visualize the difference

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
data1 = np.random.normal(50, 10, 1000)  # Mean 50, Std Dev 10
data2 = np.random.normal(60, 15, 1000)  # Mean 60, Std Dev 15

# Calculate standard deviations
std1 = np.std(data1)
std2 = np.std(data2)
print(f"Standard Deviation of Dataset 1: {std1}")
print(f"Standard Deviation of Dataset 2: {std2}")

# Visualize
plt.hist(data1, bins=30, alpha=0.5, label="Dataset 1")
plt.hist(data2, bins=30, alpha=0.5, label="Dataset 2")
plt.title("Comparison of Distributions")
plt.legend()
plt.show()
```

QUE 22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

```python
import numpy as np

# Generate data from a normal distribution
data = np.random.normal(50, 10, 1000)

# Calculate range and IQR
data_range = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {iqr}")
```

QUE 23 Implement Z-score normalization on a dataset and visualize its transformation

```python
from scipy.stats import zscore
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(50, 10, 1000)

# Perform Z-score normalization
normalized_data = zscore(data)

# Visualize
plt.hist(data, bins=30, alpha=0.5, label="Original Data")
plt.hist(normalized_data, bins=30, alpha=0.5, label="Z-score Normalized Data")
plt.title("Z-score Normalization")
plt.legend()
plt.show()
```
